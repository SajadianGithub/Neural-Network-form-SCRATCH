```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
def f(x, y):
    return x**2 + 10 * y**2
```

```python
def grad_f(x,y):
    return (np.array([2*x, 20*y]))
```

```python
X = np.arange(-3, 3, 0.01)
Y = np.arange(-3, 3, 0.01)
X, Y = np.meshgrid(X, Y)
Z = f(X, Y) 
```

```python
xk = np.array([2, 1])
seri_x = []
seri_y = []
seri_x.append(xk[0])
seri_y.append(xk[1])

for i in range (100): 
    xn = xk - 0.1 * grad_f(xk[0], xk[1])

    seri_x.append(xn[0])
    seri_y.append(xn[1])
    xk = np.copy(xn)
```

```python
plt.figure(figsize=(12,7))
plt.contour(X, Y, Z, 100, colors='black')
plt.scatter(seri_x, seri_y, )
plt.plot(seri_x, seri_y)
# for i, txt in enumerate(range(len(seri_x))):
#     plt.annotate(txt, (seri_x[i], seri_y[i]))
plt.scatter(0, 0, color = "red",)
# plt.figure().set_figwidth(100)
# plt.figure().set_figheight(5)
plt.xlim([-1, 2])
plt.ylim([-1.5, 1.5])
```

```python
xk = np.array([2, 1])
seri_x = []
seri_y = []
seri_x.append(xk[0])
seri_y.append(xk[1])

dk = 0

for i in range (100):
    dk = 0.9 * dk - 0.01 * grad_f(xk[0], xk[1]) 
    xn = xk + dk

    seri_x.append(xn[0])
    seri_y.append(xn[1])
    xk = np.copy(xn)

```

```python
plt.figure(figsize=(12,7))
plt.contour(X, Y, Z, 100, colors='black')
plt.scatter(seri_x, seri_y, )
plt.plot(seri_x, seri_y)
# for i, txt in enumerate(range(len(seri_x))):
#     plt.annotate(txt, (seri_x[i], seri_y[i]))
plt.scatter(0, 0, color = "red",)
# plt.figure().set_figwidth(100)
# plt.figure().set_figheight(5)
plt.xlim([-1, 2])
plt.ylim([-1, 1])
```

```python
xk = np.array([2, 1])
seri_x = []
seri_y = []
seri_x.append(xk[0])
seri_y.append(xk[1])
mk = 0
sk = 0

alfa = 0.1
beta_1 = 0.9
beta_2 = 0.9
for i in range (100):
    mk = beta_1 * mk + (1 - beta_1) * grad_f(xk[0], xk[1]) 
    sk = beta_2 * sk + (1 - beta_2) * np.multiply(grad_f(xk[0], xk[1]), grad_f(xk[0], xk[1]))

    mkbar = mk/(1 - ((beta_1)**(i+1)))
    skbar = sk/(1 - ((beta_2)**(i+1)))

    xn = xk - alfa*(mkbar / (skbar**0.5))

    seri_x.append(xn[0])
    seri_y.append(xn[1])
    xk = np.copy(xn)
```

```python
plt.figure(figsize=(12,7))
plt.contour(X, Y, Z, 100, colors='black')
plt.scatter(seri_x, seri_y, )
plt.plot(seri_x, seri_y)
# for i, txt in enumerate(range(len(seri_x))):
#     plt.annotate(txt, (seri_x[i], seri_y[i]))
plt.scatter(0, 0, color = "red",)
# plt.figure().set_figwidth(100)
# plt.figure().set_figheight(5)
plt.xlim([-1, 2])
plt.ylim([-1, 1])
```
